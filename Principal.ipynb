```python
# which show is the most frequently watched in México,Brazil an China:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset desde la ruta proporcionada
df = pd.read_csv('Resources/netflix_merged.csv')

# Filtrar solo las filas donde el tipo es "Movie"
df_movies = df[df['type'] == 'Movie']

# Contar cuántas películas están asociadas a cada país
# Asumimos que los países están separados por comas
country_counts = df_movies['country'].str.split(',').explode().str.strip().value_counts()

# Filtrar los tres países con más películas
top_countries = country_counts.head(3)

# Mostrar la tabla de resultados
print("Top 3 Países con más Películas en la Plataforma:")
top_countries_df = top_countries.reset_index()
top_countries_df.columns = ['País', '# de Películas']
top_countries_df['Posición'] = top_countries_df.index + 1
top_countries_df = top_countries_df[['Posición', 'País', '# de Películas']]
print(top_countries_df)

# Gráfico de barras (vertical)
plt.figure(figsize=(10, 6))
plt.bar(top_countries.index, top_countries.values, color=['lightblue', 'lightgreen', 'lightyellow'])
plt.xlabel('País')
plt.ylabel('# de Películas')
plt.title('Top 3 Países con más Películas en la Plataforma')
plt.xticks(rotation=45)
plt.show()

# Gráfico de pastel
plt.figure(figsize=(7, 7))
plt.pie(top_countries.values, labels=top_countries.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightyellow'], startangle=90)
plt.title('Distribución de Películas entre los Top 3 Países')
plt.show()

# Gráfico de barras horizontales
plt.figure(figsize=(10, 6))
plt.barh(top_countries.index, top_countries.values, color=['lightblue', 'lightgreen', 'lightyellow'])
plt.xlabel('# de Películas')
plt.ylabel('País')
plt.title('Top 3 Países con más Películas en la Plataforma')
plt.show()

# Tabla visual simple con seaborn
plt.figure(figsize=(8, 3))
sns.heatmap(top_countries_df.set_index('País').T, annot=True, cmap='Blues', cbar=False, fmt='d')
plt.title('Top 3 Países con más Películas')
plt.show()
```

### First: let's figure out wich countries produce more content: 

```python
#In this section we are listing the countries that create content: 

per_country_counts = netflix_df.set_index(["Country"])["Title"].groupby(level=0).size()
per_country_counts = per_country_counts.sort_values(ascending=False)
per_country_counts.head(10)
```

```python
#Create a dataframe to visualize into a chart the top 10 countries that create movies. 

top_countrie_movies = pd.DataFrame(per_country_counts.head(10).copy())
top_countrie_movies['index1'] = top_countrie_movies.index #Converts the index into a column value
top_countrie_movies = top_countrie_movies.reset_index(drop=True) #deletes the index since is no longer used


#imports country converter and changes the value "country" to the ISO identifier of the country
import country_converter as coco
top_countrie_movies['index1'] = top_countrie_movies['index1'].apply(lambda x: coco.convert(names=x, to='ISO3', not_found=None))
#print(top_countrie_movies)

#Creates the dataset to create polygons in the shape of countries
import geopandas as gpd
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
df_world = gpd.read_file(url)
#df_world.plot(figsize=(10,6))


#Merge the count of countryes and the polygon dataset. 
df_to_dataprint = df_world.merge(top_countrie_movies, how="right", left_on=['ADM0_ISO'], right_on=['index1'])

#creates a map with only the boundaries of the countries
ax = df_world["geometry"].boundary.plot(figsize=(20,16))

#prints into the map, the count of countries that creates content
df_to_dataprint.plot( column="index1", ax=ax, cmap='PuRd')
ax.set_title("Countries tha creates more content for nextflix")

```

```python
#Creates a bar chart to show the top countries

import numpy as np
import matplotlib.pyplot as plt 


content = df_to_dataprint['Title'] 
country = df_to_dataprint['SOVEREIGNT']  


#Creates the bar chart
bar1 = plt.bar(country,content,color='#E50914')
bar1 = plt.xticks(rotation=90)

#Set tittles to the bar chart
bar1 = plt.title('Movies and series created by country')
bar1 = plt.xlabel('Country')
bar1 = plt.ylabel('Content')
plt.show(bar1)
```

### Then: let's figure out the top countries that on average are creating the newest (premieres), and the oldest content 


```python
#In this section we are figuring out the newest and oldest years of content creation, just for curiosity: 

#We obtain the column with the year data, and then we sort it descending, to finally print the first 5 values 
Top_newest_years =  netflix_df['Available to the public (year)'].unique()
Top_newest_years = sorted(Top_newest_years, reverse=True)
print(f'The most recent years that contert were created: {Top_newest_years[:5]}')


#We obtain the column with the year data, and then we sort it ascending, to finally print the first 5 values 
Top_oldest_years =  netflix_df['Available to the public (year)'].unique()
Top_oldest_years = sorted(Top_oldest_years, reverse=False)
print(f'The oldest years when contetn were created: {Top_oldest_years[:5]}')

```

```python
# Then we create a table with the content created by year (most recent), by country


#create a dataframe only with country data
histogram = pd.DataFrame(netflix_df["Country"].unique())

#Loop to create a column with the count of content created by country by year (considering only the 5 newest years)
for years in Top_newest_years[:5]:
    df_yearfilter = netflix_df.loc[netflix_df['Available to the public (year)'] == years]
    df_yearfilter_country = df_yearfilter.set_index(["Country"])['Title'].groupby(level=0).size()
    df_filter_final = pd.DataFrame(df_yearfilter_country.copy())
    df_filter_final = df_filter_final.sort_values(by=['Title'], ascending=False)
    df_filter_final_columns = df_filter_final.rename(columns={'Title': str(f'Titles for {years}')})
    histogram = pd.merge(histogram, df_filter_final_columns, left_on=0, right_index=True)

#sort the values
histogram =  histogram.sort_values(by=['Titles for 2021'], ascending=False)

#selects only the first 5
first = histogram[:5]

#transposes the data
transposed = first.transpose()

#rename the column name
transposed.columns = transposed.iloc[0]

#delete the first row as we don't need it
transposed = transposed[1:]

#resets index
transposed = transposed.reset_index()	

#rename the column index
transposed = transposed.rename(columns={'index': 'Tittles created by year'})

#prints the chart
transposed.plot('Tittles created by year',['United States', 'India', 'United Kingdom', 'South Korea','Spain'],kind = 'line')
```

```python
#Finally we create a table with the countries with the oldest content


#create a dataframe only with country data
histogram2 = pd.DataFrame(netflix_df["Country"].unique())


#Loop to create a column with the count of content created by country by year (considering only the 5 newest years)
for years in Top_oldest_years[:5]:
    df_yearfilter = netflix_df.loc[netflix_df['Available to the public (year)'] == years]
    df_yearfilter_country = df_yearfilter.set_index(["Country"])['Title'].groupby(level=0).size()
    df_filter_final = pd.DataFrame(df_yearfilter_country.copy())
    df_filter_final = df_filter_final.sort_values(by=['Title'], ascending=False)
    df_filter_final_columns = df_filter_final.rename(columns={'Title': str(f'Titles for {years}')})
    histogram2 = pd.merge(histogram2, df_filter_final_columns, left_on=0, right_index=True, how='left')


#sort the values
histogram2 =  histogram2.sort_values(by=['Titles for 1925'], ascending=False)

#selects only the first 5
first = histogram2[:5]

#transposes the data
transposed = first.transpose()

#rename the column name
transposed.columns = transposed.iloc[0]

#delete the first row as we don't need it
transposed = transposed[1:]

#resets index
transposed = transposed.reset_index()	

#rename the column index
transposed = transposed.rename(columns={'index': 'Tittles created by year'})


#prints the chart
transposed.plot('Tittles created by year',['United States'],kind = 'line', color= '#E50914')
```
