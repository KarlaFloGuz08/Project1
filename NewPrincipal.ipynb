```python
#Monse Questions  1.-Which countries have the most movies on the platform? MONSE
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the provided path
df = pd.read_csv('Resources/netflix_merged.csv')

# Filter only rows where the type is "Movie"
df_movies = df[df['type'] == 'Movie']

# Count how many movies are associated with each country
# Assume countries are separated by commas
country_counts = df_movies['country'].str.split(',').explode().str.strip().value_counts()

# Filter the top three countries with the most movies
top_countries = country_counts.head(3)

# Bar chart (vertical) with Netflix red tone
netflix_red = '#E50914'
plt.figure(figsize=(10, 6))
plt.bar(top_countries.index, top_countries.values, color=netflix_red)
plt.xlabel('Country')
plt.ylabel('# of Movies')
plt.title('Top 3 Countries with the Most Movies on the Platform')
plt.xticks(rotation=45)
plt.show()

# Pie chart with Netflix red and black tones
plt.figure(figsize=(7, 7))
plt.pie(top_countries.values, labels=top_countries.index, autopct='%1.1f%%', colors=[netflix_red, 'darkred', 'black'], startangle=90)
plt.title('Movie Distribution Among the Top 3 Countries')
plt.show()
```

```python
#Monse Questions  2.-Who are the top actors or actresses with the most content on the platform, and in which countries is their content available?
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Resources/netflix_merged.csv')

# Clean up the `cast` column by splitting the actors and exploding the list
df['cast'] = df['cast'].fillna('')  # Fill NaN values with empty strings
df_cast = df[df['cast'] != '']  # Filter out rows without cast information
df_cast = df_cast.assign(actor=df_cast['cast'].str.split(',')).explode('actor')
df_cast['actor'] = df_cast['actor'].str.strip()  # Remove any extra whitespace

# Count the appearances of each actor
actor_counts = df_cast['actor'].value_counts().head(3)

# Get the top 3 actors and the titles of their movies/shows and countries
top_actors = actor_counts.index
top_actor_info = {
    actor: {
        'Titles': df_cast[df_cast['actor'] == actor]['title'].unique().tolist(),
        'Countries': df_cast[df_cast['actor'] == actor]['country'].dropna().unique().tolist()
    } 
    for actor in top_actors
}

# Create a DataFrame for the summary table
summary_table = pd.DataFrame({
    'Actor': top_actors,
    'Number of Movies/Shows': actor_counts.values,
    'Titles': [", ".join(info['Titles']) for info in top_actor_info.values()],
    'Countries': [", ".join(info['Countries']) for info in top_actor_info.values()]
})

# Display the summary table
print("Summary Table of Top 3 Actors with Most Content on the Platform:")
print(summary_table)

# Display the table in a formatted view (if using Jupyter Notebook)
from IPython.display import display
display(summary_table)

# Horizontal bar chart for Number of Movies/Shows
plt.figure(figsize=(10, 6))
bars = plt.barh(summary_table['Actor'], summary_table['Number of Movies/Shows'], color='#E50914')  # Netflix red color
plt.xlabel('Number of Movies/Shows')
plt.title('Top 3 Most Frequent Actors on the Platform')

# Add data labels with the number of shows for each bar
for index, value in enumerate(summary_table['Number of Movies/Shows']):
    plt.text(value + 0.5, index, str(value), va='center', ha='left', fontsize=10, color='black')

plt.gca().invert_yaxis()  # Invert y-axis to show the actor with the most appearances at the top
plt.show()
```

```python
#In this section we are listing the countries that create content: 

per_country_counts = netflix_df.set_index(["Country"])["Title"].groupby(level=0).size()
per_country_counts = per_country_counts.sort_values(ascending=False)
per_country_counts.head(10)
```

```python
#Create a dataframe to visualize into a chart the top 10 countries that create movies. 

top_countrie_movies = pd.DataFrame(per_country_counts.head(10).copy())
top_countrie_movies['index1'] = top_countrie_movies.index #Converts the index into a column value
top_countrie_movies = top_countrie_movies.reset_index(drop=True) #deletes the index since is no longer used


#imports country converter and changes the value "country" to the ISO identifier of the country
import country_converter as coco
top_countrie_movies['index1'] = top_countrie_movies['index1'].apply(lambda x: coco.convert(names=x, to='ISO3', not_found=None))
#print(top_countrie_movies)

#Creates the dataset to create polygons in the shape of countries
import geopandas as gpd
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
df_world = gpd.read_file(url)
#df_world.plot(figsize=(10,6))


#Merge the count of countryes and the polygon dataset. 
df_to_dataprint = df_world.merge(top_countrie_movies, how="right", left_on=['ADM0_ISO'], right_on=['index1'])

#creates a map with only the boundaries of the countries
ax = df_world["geometry"].boundary.plot(figsize=(20,16))

#prints into the map, the count of countries that creates content
df_to_dataprint.plot( column="index1", ax=ax, cmap='GnBu')
ax.set_title("Countries tha creates more content for nextflix")

```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# Load the dataset
netflix_path = Path("Resources/netflix_titles.csv")
netflix_df = pd.read_csv(netflix_path, encoding="utf-8")

# Define family-friendly ratings
family_friendly_ratings = ['TV-G', 'TV-Y', 'TV-Y7', 'TV-PG']

# Count total number of titles
total_titles = netflix_df.shape[0]

# Count family-friendly titles
family_friendly_count = netflix_df[netflix_df['rating'].isin(family_friendly_ratings)].shape[0]
non_family_friendly_count = total_titles - family_friendly_count

# Prepare data for plotting
labels = ['Family-Friendly Content', 'Non-Family-Friendly Content']
sizes = [family_friendly_count, non_family_friendly_count]
colors = ['#e50914', '#ff6f61']  # Netflix Red and a vibrant Coral color
explode = (0.1, 0)  # explode the 1st slice (Family-Friendly)

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140, 
        textprops=dict(color="white"))  # White text for better contrast
plt.title('Distribution of Family-Friendly vs Non-Family-Friendly Content on Netflix', fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
from pathlib import Path

# Load the dataset
netflix_path = Path("Resources/netflix_titles.csv")
netflix_df = pd.read_csv(netflix_path, encoding="utf-8")

# Define family-friendly ratings
family_friendly_ratings = ['TV-G', 'TV-Y', 'TV-Y7', 'TV-PG']

# Count total number of titles
total_titles = netflix_df.shape[0]

# Count family-friendly titles
family_friendly_count = netflix_df[netflix_df['rating'].isin(family_friendly_ratings)].shape[0]

# Calculate percentage of family-friendly content
family_friendly_percentage = (family_friendly_count / total_titles) * 100

# Count mature content (TV-MA)
mature_count = netflix_df[netflix_df['rating'] == 'TV-MA'].shape[0]

# Calculate percentage of mature content
mature_percentage = (mature_count / total_titles) * 100

# Display the results
print(f"Total Titles: {total_titles}")
print(f"Family-Friendly Titles: {family_friendly_count}")
print(f"Percentage of Family-Friendly Content: {family_friendly_percentage:.2f}%")
print(f"Mature Titles (TV-MA): {mature_count}")
print(f"Percentage of Mature Content: {mature_percentage:.2f}%")
```
